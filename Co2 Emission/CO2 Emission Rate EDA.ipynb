## Visualization of the DataFrame :

```python
# 1. CO2 Emissions by Year :

plt.figure(figsize=(10, 6))
plt.bar(EHG_df['year'], EHG_df['total_emissions_MtCO2e'], color='skyblue')
plt.title('CO2 Emissions by Year')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (in metric tons)')
plt.xticks(EHG_df['year'])
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# 2. Distribution of Years :

plt.figure(figsize=(10, 6))
sns.countplot(data=EHG_df, x='year')
plt.title('Distribution of Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Bar Plot and Line Plot Subplots (Comparing different categories and trends over time) :


fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))

# Bar Plot
axes[0].bar(EHG_df['parent_entity'], EHG_df['total_emissions_MtCO2e'])
axes[0].set_xlabel('Parent Entity')
axes[0].set_ylabel('Total Emissions (MtCO2e)')
axes[0].set_title('Total Emissions by Parent Entity')

# Line Plot
axes[1].plot(EHG_df['year'], EHG_df['total_emissions_MtCO2e'], marker='o', linestyle='-')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Total Emissions (MtCO2e)')
axes[1].set_title('Total CO2 Emissions Over Time')

plt.tight_layout()
plt.show()
```

```python
# 4. Scatter Plot Matrix (Exploring relationships between multiple variables) :

# Selecting relevant numeric columns for the scatter plot matrix
numeric_colm = ['production_value', 'product_emissions_MtCO2', 'flaring_emissions_MtCO2', 
                'venting_emissions_MtCO2', 'own_fuel_use_emissions_MtCO2', 
                'fugitive_methane_emissions_MtCO2e', 'fugitive_methane_emissions_MtCH4', 
                'total_operational_emissions_MtCO2e', 'total_emissions_MtCO2e']

# Creating a pairplot
sns.pairplot(EHG_df[numeric_colm])
plt.suptitle('Pairplot of Numeric Variables', y=1.02)
plt.show()
```

```python
# 5. Total Emissions by Commodity :

plt.figure(figsize=(10, 6))
sns.boxplot(x='commodity', y='total_emissions_MtCO2e', data=EHG_df)
plt.xlabel('Commodity')
plt.ylabel('Total Emissions (MtCO2e)')
plt.title('Total Emissions by Commodity')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Total CO2 Emissions Over Time :

plt.figure(figsize=(10, 6))
plt.plot(EHG_df['year'], EHG_df['total_emissions_MtCO2e'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Emissions (MtCO2e)')
plt.title('Total CO2 Emissions Over Time')
plt.grid(True)
plt.show()
```

```python
# 7. Production Value by Commodity :

plt.figure(figsize=(10, 6))
plt.bar(EHG_df['commodity'], EHG_df['production_value'])
plt.xlabel('Commodity')
plt.ylabel('Production Value')
plt.title('Production Value by Commodity')
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. Scatter Plot of Production Value vs Total Emissions :

plt.figure(figsize=(10, 6))
plt.scatter(EHG_df['production_value'], EHG_df['total_emissions_MtCO2e'])
plt.xlabel('Production Value')
plt.ylabel('Total Emissions (MtCO2e)')
plt.title('Production Value vs Total Emissions')
plt.grid(True)
plt.show()
```

```python
# 9. Distribution of Total CO2 Emissions :

plt.figure(figsize=(10, 6))
plt.hist(EHG_df['total_emissions_MtCO2e'], bins=20, color='skyblue')
plt.xlabel('Total Emissions (MtCO2e)')
plt.ylabel('Frequency')
plt.title('Distribution of Total Emissions')
plt.grid(True)
plt.show()
```

```python
# 10.Total Emissions Distribution by Parent Type :

plt.figure(figsize=(10, 6))
sns.boxplot(x='parent_type', y='total_emissions_MtCO2e', data=EHG_df)
plt.xlabel('Parent Type')
plt.ylabel('Total Emissions (MtCO2e)')
plt.title('Total Emissions Distribution by Parent Type')
plt.show()
```

```python
# 11. Proportion of Total Emissions by Parent Entity :

plt.figure(figsize=(8, 8))
plt.pie(EHG_df.groupby('parent_entity')['total_emissions_MtCO2e'].sum(), labels=EHG_df['parent_entity'].unique(), autopct='%1.1f%%')
plt.title('Total Emissions by Parent Entity')
plt.show()
```

```python
# 12. Correlation Matrix of Numeric Variables :

# Select only numeric columns for correlation calculation
numeric_df = EHG_df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10, 8))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
